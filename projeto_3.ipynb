```python
for loja in base['Outlet_Identifier'].unique():
  dados_tamanho_loja = base.loc[base['Outlet_Identifier'] == loja, 'Outlet_Size']
  if dados_tamanho_loja.isnull().sum() > 0:
    print('loja:', loja, '\nqtd nulos:', dados_tamanho_loja.isnull().sum(), '\n', dados_tamanho_loja.value_counts(), '\n')
```

```python
items_peso_nulo = base.loc[base['Item_Weight'].isnull(), 'Item_Identifier'].unique()
```

```python
base[base['Item_Identifier'].isin(items_peso_nulo)].sample(5)
```

```python
# verificar se todos os items estao com o mesmo peso na base, assim podendo usar esse peso para imputar os valores faltantes
# como todos os items estao "Padrinizados", irei substituir os pesos dos itens ja iguals que ja foram registrados na base
for item in items_peso_nulo:
  qtd_unicos = len(base.loc[base['Item_Identifier'] == item, 'Item_Weight'].unique())
  if qtd_unicos > 2:
    print(item)
```

```python
# vimos acima, todos os pesos estao positivos, entao podemos usar o -1 para identifica-los
base['Item_Weight'].fillna(-1, inplace=True)
```

```python
for item in items_peso_nulo:
  valor_impute = base.loc[base['Item_Identifier'] == item, 'Item_Weight'].max()
  base.loc[(base['Item_Identifier'] == item) & (base['Item_Weight'] == -1) , 'Item_Weight'] = valor_impute
```

```python
base.loc[base['Item_Weight'] == -1, 'Item_Weight'] = base['Item_Weight'].median()
```

```python
base.isnull().sum()[base.isnull().sum() > 0] # pesos dos itens nulos ja tratados, falta penas o tamanho das lojas
```

```python
base.fillna('Nulo').groupby(['Outlet_Type', 'Outlet_Size'])[['Item_Identifier']].count()
```

```python
# Vamos adotar algumas premissas nesse momento, primeira delas é considerar que as Grocery Store
# são todas Small (que faz sentido, mercearias geralmente sao pequenas mesmo)
# irei fazer os testes, se nao der certo irei usar o fator aleatoriedade

base.loc[(base['Outlet_Type'] == 'Grocery Store') & (base['Outlet_Size'].isnull()), 'Outlet_Size'] = 'Small'
base.loc[base['Outlet_Size'].isnull(), 'Outlet_Size'] = np.random.choice(base['Outlet_Size'].dropna().unique(), size=base.loc[base['Outlet_Size'].isnull(), 'Outlet_Size'].shape[0])
```

```python
base_predicao = base.drop(['lucro', 'Item_Identifier', 'Outlet_Identifier', 'Item_Weight'], axis=1)
base_predicao['Outlet_Establishment_Year'] = base_predicao['Outlet_Establishment_Year'].astype('object')
```

```python
target = base_predicao.pop('Item_Outlet_Sales')
```

```python
import sweetviz as sv

advert_report = sv.analyze(base)
advert_report.show_html('bigmart.html')
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
sc.fit_transform(base_predicao.select_dtypes(exclude='object'))
```

```python
cat = pd.get_dummies(base_predicao.select_dtypes('object'), drop_first=True)
num = pd.DataFrame(sc.fit_transform(base_predicao.select_dtypes(exclude='object')), columns=sc.get_feature_names_out())
```

```python
base_final_predicao = pd.concat((num, cat), axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, mean_absolute_error, r2_score
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV, KFold

modelo_linear = GridSearchCV(LinearRegression(), cv=KFold(5), param_grid={})
modelo_ensaboado = GridSearchCV(XGBRegressor(), cv=KFold(5), param_grid={})

# ajustar parametros para criar um modelo mais otimizado
```

```python
modelo_linear.fit(base_final_predicao, target)
modelo_ensaboado.fit(base_final_predicao, target)
```

```python
display(mean_absolute_error(target, modelo_linear.predict(base_final_predicao)))
display(mean_absolute_error(target, modelo_ensaboado.predict(base_final_predicao)))
```

```python
display(modelo_linear.score(base_final_predicao, target))
display(modelo_ensaboado.score(base_final_predicao, target))
```

```python
beta_linear = modelo_linear.best_estimator_.coef_
beta_linear = pd.DataFrame(zip(base_final_predicao.columns, beta_linear), columns=['colunas', 'beta'])
beta_linear.sort_values(by='beta')
```

```python
beta_ensaboado = modelo_ensaboado.best_estimator_.feature_importances_
beta_ensaboado = pd.DataFrame(zip(base_final_predicao.columns, beta_ensaboado), columns=['colunas', 'beta'])
beta_ensaboado.sort_values(by='beta')
```
